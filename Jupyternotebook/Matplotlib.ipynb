```python
x = numpy.arange(0,5,0.1)
y = x**3
plt.plot(x,y,color="yellow",linewidth="5")
plt.show()
```

```python
#********************** Labeling along x and y and giving title to plot ***************************
x = numpy.arange(0,5,0.1)
y = x**3
y1 = x**2
plt.plot(x,y,color="green",linewidth="5",label="x^3")
plt.plot(x,y1,"r-.",label="x^2")
plt.axis([2,10,2,100]) # axis is used to change the axis values according to us, takes 0 or 1 positional
# arguments and first argument should be iterable like [min_x,max_x,min_y,max_y]
plt.grid() # grid function is used to create grids
plt.text(4,60,"(4,60)",fontsize="12")
plt.ylabel("y")
plt.xlabel("x")
plt.legend() #for showing the labels given to different lines in the x Vs y plot
plt.title("x Vs x^3 graph")
plt.show()
```

```python
year = [2012,2013,2014,2015,2016,2017]
salary = [12,13,14,17,19,20]
population = [100,120,180,250,300,370]
plt.scatter(year,salary,s=population,color=["red","green","blue","yellow","pink","orange"],alpha=0.5) 
# parameters should always have their name as specified in documentation else will raise error as changing
# name from "s" to "size" raises error
# if s=constant then all the bubbles have same size
# but if (s = some array ) then changes size of bubble according to array values like s= population
# if color = "color_name" then changes color of all the bubbles to that color
# but if color=array containing different colors then each bubble shows that color 
# the size of the color array or size array should be equals to the size of x and y 
# alpha is used to adjust the transparency of the bubble's color,
#  it's default value = 1and it is in the range of 0-1
plt.title("Population scattering")
plt.xlabel("salary")
plt.ylabel("year")
plt.grid()
```

```python
import matplotlib.pyplot as plt
import numpy
year = [2012,2013,2014,2015,2016,2017]
salary = [12,13,14,17,19,20]
population = [100,120,180,250,300,370]
c = numpy.arange(len(year))
print(c)
c[0] = c[2] #color of 0th index bubble and 2nd index bubble will be same
plt.scatter(year,salary,s=population,c=c,alpha=1,edgecolors="black",marker="^") # c is deciding various colors 
# alpha is for intensity or transparency of the bubbles by default it's value is 1,
# value of alpha is in the range of 0-1
# edgecolors argument changes the edge of color to specified color 
plt.grid()
plt.text(2015,17,"color:aquamarine") # text insertion after griding(obseriving grid then texting)
s = str(population[2]) 
plt.text(year[2]+0.1,salary[2]+0.1,s) # text insertion directly
# adding 0.1 in year and salary to avoid the overlapping of text with markers
plt.show()
```

```python
#*************************** Pie chart ********************************
sizes = [3,6,9,12]
labels = ["Grey","Orange","Yellow","Hotpink"]
colors=["grey","orange","yellow","hotpink"]
plt.title("Various colors chart")
explode = [0.1,0,0,0]
plt.pie(sizes,colors=colors,labels=labels,autopct="%.2f",explode=explode,counterclock=True,startangle=90) 
#using autopct we can write over each section 
# and in autopct we pass a function (we can create using def and can pass to it)

plt.axis("equal") #it divides 4 quadrants equally then divides each section on different quadrants
#(either in only 1 or takes two or more quadrant as per requirement)

plt.show()
```
